```python
import matplotlib.pyplot as plt
import pandas as pd


dados = pd.read_csv("C:/Users/marce/ansel/Desktop/cod/athlete_events.csv")

dados.hist(column= "Age" , bins=10) #bins é barras
plt.show()

"""calculo de um historograma:
pegamos o minímo e máximo: nesse caso, idades. vamos supor que as idades é min 10 e max 90
fazemos o calculo de 90 - 10 = 80, como botamos 10 colunas, devemos dividir 80 / 10 = 8
ou seja, pegamos do ponto min (10) somamos + 8 = 18, cerca de 20mil atletas tinham de 10 a 18 anos.
após isso, temos que 18 + 8 = 26, cerca de mais de 160 mil atletas tiveram essa idade"""
```
