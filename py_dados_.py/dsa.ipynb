```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Qual o Total de Vendas Por Data do Pedido?Demonstre o resultado através de um gráfico de barras

dados = pd.read_csv("C:/Users/marce/Downloads/dados/dataset.csv") 

totaldata = dados.groupby("Data_Pedido") ["Valor_Venda"].sum()

totaldata


plt.figure(figsize= (20, 6))
totaldata.plot(x= "Data_Pedido" , y= "Valor_Venda" , color= "Blue")
plt.title('Venda por data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
dados = pd.read_csv("C:/Users/marce/Downloads/dados/dataset.csv") 

vendas_estado = dados.groupby("Estado")["Valor_Venda"].sum()
vendas_estado.sort_values(ascending= False) #descobrindo os maiores
top10 = vendas_estado[vendas_estado>= 68194.6700]
top10


plt.figure(figsize= (10,10))
top10.plot(x= "Estado", y= "Valor_Venda", color = "Red")
plt.title('TOP 10 ESTADOS QUE MAIS VENDEM')
plt.show()
```
